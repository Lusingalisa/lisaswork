```python
Cassava=cassava.drop_duplicates()
cassava.shape
```

```python
Cassava.head(10)
```

```python
ContCass=Cassava.select_dtypes(include=['float64','int64'])
```

```python
ContCass.head()
```

```python
ContinuousCols=['Plants_harvested',	'No_bigtubers',	'Weigh_bigtubers',	'No_mediumtubers', 'Weight_mediumtubers', 'No_smalltubers',	'Weight_smalltubers', 'Totaltuberno']

for predictor in ContinuousCols:
    Cassava.plot.scatter(x= predictor, y= 'Sesn', figsize=(10,5), title= predictor  +  "AGAINST"  +  'Sesn')
```

```python
ContinuousVariables=['Plants_harvested',	'No_bigtubers',	'Weigh_bigtubers',	'No_mediumtubers', 'Weight_mediumtubers', 'No_smalltubers',	'Weight_smalltubers', 'Totaltuberno', 'Sesn']
Correlationdata=cassava[ContinuousVariables].corr()
Correlationdata
```

```python
Correlationdata['Sesn'][abs(Correlationdata['Sesn'])>0.5]
```

- There is a moderate correlation between the plants harvested and the season it is
- This means the type of season affects the number of plants harvested

```python
ContCols=['No_bigtubers']
for predictor in ContCols:
    Cassava.plot.scatter(x=predictor, y='Sesn', figsize=(10,5), title=predictor+ "VS" + 'Sesn')
```

```python
contvariables=['Sesn','No_bigtubers']
correlationdata=Cassava[contvariables].corr()
correlationdata
```
